# Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas=pd.read_csv('gas_prices.csv')
gas
```

# Line Graph

```python
plt.figure(figsize=(8,6))
plt.title('Gas Prices over time in USD')
plt.plot(gas['Year'],gas['USA'],'b.-')
plt.plot(gas['Year'],gas['Canada'],'r.-')
plt.plot(gas['Year'],gas['South Korea'],'g.-')
# Take value from every 3 rd year
plt.xticks(gas['Year'][::3])
plt.xlabel('Year')
plt.ylabel('US$')
# plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,10))
# Below snippet plots all the countries prices
for country in gas:
    if country != 'Year':
        plt.plot(gas['Year'],gas[country],'.-',label=country)
plt.legend()
plt.xlabel("Years")
plt.ylabel('Price in US$')
plt.savefig("Gas_prices_plot.png",dpi=300)
plt.show()
```

# Load FIFA Data

```python
fifa=pd.read_csv('fifa_data.csv')
fifa.head()
```

# Histograms

```python
bins=[40,50,60,70,80,90,100]
plt.hist(fifa['Overall'],bins=bins,color='#abcdef')
plt.xticks(bins)
plt.ylabel('Number of Players')
plt.xlabel('Distribution of Player Skills in FIFA 2018')
plt.show()
```

# Pie Charts

```python
left=fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
labels=['Left','Right']
colors=['#abcdef','#fedcba']
plt.pie([left,right],labels=labels,colors=colors,autopct='%.2f')
plt.show()
```

```python
fifa['Weight']=[int(x.strip('lbs')) if type(x)==str else x for x in fifa['Weight']]
light=fifa.loc[fifa['Weight'] < 125].count()[0]
light_medium=fifa.loc[(fifa['Weight'] >= 125) & (fifa['Weight'] < 150)].count()[0]
medium=fifa.loc[(fifa['Weight'] >=150) & (fifa['Weight']< 175)].count()[0]
heavy_medium=fifa.loc[(fifa['Weight'] >= 175) & (fifa['Weight'] < 200)].count()[0]
heavy=fifa.loc[(fifa['Weight'] >= 200)].count()[0]
```

```python
weights=[light,light_medium,medium,heavy_medium,heavy]
labels=['Under 125','Between 125 and 150',
        'Between 150 and 175','Between 175 and 200','Over 200']
plt.style.use('ggplot')
explode=(0.4,.2,0,0,.4)
plt.pie(weights,labels=labels,autopct='%.2f',pctdistance=0.5,
        explode=explode)
plt.show()
```

```python
barcelona=fifa.loc[fifa['Club']=='FC Barcelona']['Overall']
madrid=fifa.loc[fifa['Club']=='Real Madrid']['Overall']
revs=fifa.loc[fifa['Club']=='New England Revolution']['Overall']
```

```python
plt.style.use('default')
labels=['FC Barcelona','Real Madrid','New England Revlution']
plt.boxplot([barcelona,madrid,revs],labels=labels)
plt.title("Football Team Comparision")
plt.ylabel("FIFA Overall Rating")
plt.show()
```

# Scatter Plots

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.style.use('seaborn')
data=pd.read_csv('views.csv')
```

```python
view_count=data['view_count']
likes=data['likes']
ratio=data['ratio']
```

```python
plt.title("Trending Youtube videos")
plt.xscale('log')
plt.yscale('log')

# plt.figure().set_figheight(5)
# plt.figure().set_figwidth(5)

plt.xlabel("View Count")
plt.ylabel("Total Likes")
plt.tight_layout()
plt.scatter(view_count,likes,edgecolor='black',c=ratio,
           cmap='summer',linewidth=1,alpha=0.5)
cbar=plt.colorbar()
cbar.set_label('Like/Dislike Ratio')
```
